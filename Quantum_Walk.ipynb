```python
qc = QuantumCircuit(6, name = "Quantum Walker")
```

```python
qc.h([4,5])
qc.z([4,5])
qc.cz(4,5)
qc.h([4,5])
qc.draw()
```

```python
def shift_op(circuit):
    for i in range(0,4):
        circuit.x(4)
        if i%2==0:
            circuit.x(5)
        circuit.ccx(4,5,i)

shift_op(qc)
qc_gate = qc.to_instruction()
qc.draw()
```
